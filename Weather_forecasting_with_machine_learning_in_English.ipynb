````python
def clean_and_convert(df:pd.DataFrame,col_replace:list):
    for col in col_replace:
        if df[col].dtype == 'object': # To be more sure of the data type, if it were an object
            if df[col].astype(str).str.contains(',').any(): # If it was separated by a comma
                df[col] = df[col].str.replace(",",'.').astype('float32') # Convert to float
                print(f"{col}.......Change type from Object to Float32........OK")
                continue
            if df[col].astype(str).str.contains('/').any():
                df[col] = pd.to_datetime(df[col], format='%d/%m/%Y') # If it was separated by /, it would be converted to a datetime type.
                print(f"{col}.......Change type from Object to DateTime........OK")
                continue
            if df[col].astype(str).str.contains('.').any(): # If it was separated by a dot
                df[col] = df[col].str.split('.').str[0].astype('int32') # Convert to int
                print(f"{col}.......Change type from Object to Int32........OK")
                continue

        if df[col].dtype == 'float64': # To reduce data storage volume
            df[col] = df[col].astype("float32")
            print(f"{col}.......Change type from Float64 to Float32........OK")
    return df

col_replace = [
    'Date', 'Time', 'CO(GT)', 'PT08.S1(CO)', 'C6H6(GT)', 
    'PT08.S2(NMHC)', 'NOx(GT)', 'PT08.S3(NOx)', 'NO2(GT)', 
    'PT08.S4(NO2)', 'PT08.S5(O3)', 'T', 'RH', 'AH'
]
df=clean_and_convert(df=df,col_replace=col_replace)
````

````python
# Delete completely empty records
def drop_record_all_nans(df:pd.DataFrame):
    return df.dropna(how="all",axis=0,inplace=False)
def fillna_with_mean(df:pd.DataFrame):
    for col in df.columns:
        null_count = df[col].isnull().sum()
        if null_count < 2000:  # If the number of vacancies is less than 2000
            df[col].fillna(df[col].mean(),inplace=True)
    return df
````

### VIF (Variance Inflation Factor) Simply Explained:  
**VIF** is a number that tells you how much your **independent features** (e.g., X variables) resemble each other!  

- **Example:** Imagine your friends copying each other's exam answers. If you ask them, "Why did you choose this answer?" they both say the same thing. This means redundant information!  

---

### Interpreting VIF:  
- **1 = Perfect** (no similarity between features)  
- **1–5 = Acceptable**  
- **5–10 = Suspicious** (somewhat similar)  
- **Above 10 = Dangerous** (features are too similar, like twins!)  

---

### Why Does It Matter?  
When features are **too similar**:  
- The model gets confused about which one truly matters!  
- Results become unreliable!  

---

### Solution:  
If VIF > 10, remove one of the similar features!  



````python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculate_vif(X):
    vif = pd.DataFrame()
    vif["Feature"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif

features_option =  [
    'C6H6(GT)',     
    'PT08.S1(CO)',  
    'PT08.S5(O3)',  
    'NOx(GT)',     
    'PT08.S3(NOx)'  
]
vif_result = calculate_vif(df[features_option])
print(vif_result)
````

````python
# TODO : Import the required libraries.

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor 
from sklearn.svm import SVR
````

````python
# TODO: Train and evaluate the four initial models
# List of models
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(random_state=101),
    "Random Forest": RandomForestRegressor(random_state=101),
    "Gradient Boosting": GradientBoostingRegressor(random_state=101),
    "SVR": SVR(),
    "XGBoost": XGBRegressor(random_state=101, verbosity=0)
}

# Dictionary to store results
results = {}

#Training and evaluating each model
for name, model in models.items():
    # Model training
    model.fit(X_train, y_train)
    
    # Prediction on training and test data
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    # Calculation of criteria
    train_mae = mean_absolute_error(y_train, y_train_pred)
    train_mse = mean_squared_error(y_train, y_train_pred)
    train_r2 = r2_score(y_train, y_train_pred)
    
    test_mae = mean_absolute_error(y_test, y_test_pred)
    test_mse = mean_squared_error(y_test, y_test_pred)
    test_r2 = r2_score(y_test, y_test_pred)
    
    # Save the results
    results[name] = {
        "Train MAE": train_mae,
        "Train MSE": train_mse,
        "Train R²": train_r2,
        "Test MAE": test_mae,
        "Test MSE": test_mse,
        "Test R²": test_r2
    }

# Show results
for model, metrics in results.items():
    print(f"\n{model}:")
    print(f"Train R²: {metrics['Train R²']:.3f} | Test R²: {metrics['Test R²']:.3f}")
    print(f"Train MAE: {metrics['Train MAE']:.3f} | Test MAE: {metrics['Test MAE']:.3f}")
````

````python
# TODO : Draw the Learning Curve of each model.
from sklearn.model_selection import learning_curve
import numpy as np

def plot_learning_curve(model, X, y, cv=5):
    train_sizes, train_scores, test_scores = learning_curve(
        model, X, y, cv=cv, scoring='r2', 
        train_sizes=np.linspace(0.1, 1.0, 5),
        n_jobs=-1
    )
    
    plt.figure(figsize=(10, 6))
    plt.plot(train_sizes, np.mean(train_scores, axis=1), 'o-', label="Train Score")
    plt.plot(train_sizes, np.mean(test_scores, axis=1), 'o-', label="Test Score")
    plt.title(f"Learning Curve ({model.__class__.__name__})")
    plt.xlabel("Training Examples")
    plt.ylabel("R² Score")
    plt.legend()
    plt.grid(True)
    plt.show()

# Plotting graphs for models (e.g. Gradient Boosting and Random Forest)
plot_learning_curve(RandomForestRegressor(), X_train, y_train)
plot_learning_curve(GradientBoostingRegressor(), X_train, y_train)
plot_learning_curve(LinearRegression(),X_train,y_train)
plot_learning_curve(XGBRegressor(),X_train,y_train)
````

````python
# Define parameter search space for Gradient Boosting Regressor
param_dist_gbr = {
    'n_estimators': [20, 50, 100],          # Number of boosting stages
    'learning_rate': [0.01, 0.05, 0.07],    # Shrinkage factor
    'max_depth': [2, 4],                     # Maximum tree depth
    'min_samples_split': [3, 5, 7],          # Minimum samples to split
    'min_samples_leaf': [2, 3, 6],           # Minimum samples at leaf
    'subsample': [0.5, 0.7, 0.9],           # Fraction of samples used
    'max_features': ['sqrt', 'log2', None],  # Features considered for splits
    'loss': ['squared_error', 'huber', 'absolute_error']  # Loss functions
}

# Initialize base model
gbr_model = GradientBoostingRegressor(random_state=42)

# Configure randomized search
random_search_gbr = RandomizedSearchCV(
    estimator=gbr_model,
    param_distributions=param_dist_gbr,
    n_iter=50,                      # Number of parameter combinations to try
    scoring="neg_mean_squared_error", # Optimization metric
    cv=5,                           # 5-fold cross-validation
    verbose=3,                      # Detailed logging
    n_jobs=-1,                      # Use all CPU cores
    random_state=42                 # Reproducibility
)

# Execute search on training data
print("Starting randomized search...")
random_search_gbr.fit(X_train, y_train)
print("Search completed!")

# Evaluate best model
best_model = random_search_gbr.best_estimator_
best_model.fit(X_train, y_train)  # Retrain on full training set

# Calculate performance metrics
test_r2 = best_model.score(X_test, y_test)
train_r2 = best_model.score(X_train, y_train)

# Print results
print(f'''
{'='*60}
Best Parameters: {random_search_gbr.best_params_}
Test R² Score: {test_r2:.4f}
Train R² Score: {train_r2:.4f}
{'='*60}
''')

# Additional recommended evaluations:
y_test_pred = best_model.predict(X_test)
print("Sample Test Predictions vs Actual:")
print(pd.DataFrame({'Actual': y_test[:5], 'Predicted': y_test_pred[:5]}))
````
