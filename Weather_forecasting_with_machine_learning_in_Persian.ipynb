###  نتیحه ارزیابی در خصوص اعداد -200:

<p>عدد -200 در دیتافریم شما احتمالاً نشانگر داده‌های گمشده szیا داده‌های نامعتبر است. (Missing Values)  این عدد به‌صورت مصنوعی و غیرواقعی در دیتاست قرار داده شده تا جای خالی داده‌های واقعی را پر کند</p>
---
<p>عدد -200 در مجموعه داده‌های زیست‌محیطی و مشابه آن، معمولاً به عنوان یک مقدار نشانه‌گذاری شده برای داده‌های مفقود یا نادرست استفاده می‌شود. این مقدار به صورت قراردادی انتخاب می‌شود و ممکن است در مستندات مجموعه داده به عنوان کد برای نشان دادن عدم حضور داده واقعی ذکر شده باشد. در اینجا به چند منبع و دلیل برای استفاده از چنین عددی می‌پردازیم<p>
---

### ۱. **چرا -200 وجود دارد؟**
- **جایگزین داده‌های گمشده**:  
  ممکن است سنسورها در زمان‌هایی داده‌ها را ثبت نکرده باشند و برای پر کردن جای خالی، عددی مانند `-200` وارد شده است.
- **خطا در ثبت داده**:  
  احتمالاً سنسورها در شرایط خاصی (مثلاً دمای خیلی پایین یا بالا) قادر به ثبت داده نبوده‌اند و `-200` را برگردانده‌اند.
---

### 2. **مستندات مجموعه داده**
### اغلب مجموعه‌های داده علمی و زیست‌محیطی دارای مستنداتی هستند که توضیح می‌دهند چگونه داده‌ها جمع‌آوری شده و چگونه مقادیر خاصی به عنوان نشانه‌گذاری استفاده می‌شوند. این مستندات می‌توانند اطلاعات مفیدی در مورد معنای عدد -200 ارائه دهند.
____________________________________________________________________________________________________________________________________________
### 3. **کاربردهای عملی**
### در برخی از مطالعات، اعداد خاصی مانند -200 به عنوان کد استفاده می‌شوند تا نشان دهند که داده‌ی واقعی در دسترس نیست. این امر برای تحلیل‌گران داده و دانشمندان به منظور مدیریت و پردازش داده‌های مفقود مهم است.
____________________________________________________________________________________________________________________________________________
### 4. **تجربیات مشابه**
###  در بسیاری از مجموعه‌های داده زیست‌محیطی و آب و هوا، استفاده از مقادیر نشانه‌گذاری همچون -200 برای نشان دادن داده‌های مفقود یا نادرست رایج است. این مقدار به طور خاص نشان‌دهنده عدم دقت یا عدم وجود داده واقعی است.
____________________________________________________________________________________________________________________________________________

### 5. **مشکلات عدد -200**

````python
def clean_and_convert(df:pd.DataFrame,col_replace:list):
    for col in col_replace:
        if df[col].dtype == 'object': # برای اطمینان بیشتر از تایپ دیتا که اگر ابجکت بود
            if df[col].astype(str).str.contains(',').any(): # اگر با ویرگول جدا شده بود
                df[col] = df[col].str.replace(",",'.').astype('float32') # تبدیل شه به float
                print(f"{col}.......Change type from Object to Float32........OK")
                continue
            if df[col].astype(str).str.contains('/').any():
                df[col] = pd.to_datetime(df[col], format='%d/%m/%Y') # اگر با / جدا شده بود تایپ دیتاتایم تبدیل شه
                print(f"{col}.......Change type from Object to DateTime........OK")
                continue
            if df[col].astype(str).str.contains('.').any(): # اگر با نقطه جدا شده بود
                df[col] = df[col].str.split('.').str[0].astype('int32') # تبدیل شه به int
                print(f"{col}.......Change type from Object to Int32........OK")
                continue

        if df[col].dtype == 'float64': # برای کم کردن حجم ذخیره سازی داده ها
            df[col] = df[col].astype("float32")
            print(f"{col}.......Change type from Float64 to Float32........OK")
    return df

col_replace = [
    'Date', 'Time', 'CO(GT)', 'PT08.S1(CO)', 'C6H6(GT)', 
    'PT08.S2(NMHC)', 'NOx(GT)', 'PT08.S3(NOx)', 'NO2(GT)', 
    'PT08.S4(NO2)', 'PT08.S5(O3)', 'T', 'RH', 'AH'
]
df=clean_and_convert(df=df,col_replace=col_replace)
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 1. انتخاب ستون‌های عددی
numeric_cols = df.select_dtypes(include=['int32', 'float32']).columns.tolist()

# 2. تنظیمات بصری (استفاده از استایل موجود)
plt.style.use('seaborn-v0_8')  # یا 'ggplot', 'classic', etc.
plt.rcParams['figure.figsize'] = (18, 12)
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

# 3. بررسی توزیع داده‌ها با هیستوگرام و KDE
n_cols = 4
n_rows = int(np.ceil(len(numeric_cols) / n_cols))

fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 5*n_rows))
axes = axes.flatten()

for i, col in enumerate(numeric_cols):
    sns.histplot(
        df[col], 
        kde=True, 
        color=colors[i%4], 
        bins=30, 
        ax=axes[i]
    )
    axes[i].set_title(f'explaination {col}\n(skew: {df[col].skew():.2f})', fontsize=12)
    axes[i].axvline(df[col].mean(), color='red', linestyle='--', label='mean')
    axes[i].axvline(df[col].median(), color='green', linestyle='-.', label='median')
    axes[i].legend()

# پاکسازی subplot‌های اضافی
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
````

````python
def plot_scatter_separately(df: pd.DataFrame, target_variable: str, sensors: list):
    # لیست سنسورها برای مقایسه


    plt.figure(figsize=(20, 15))
    for i, sensor in enumerate(sensors, 1):
        plt.subplot(3, 2, i)
        sns.scatterplot(x=df[sensor], y=df[target_variable], alpha=0.5, color='blue')
        plt.title(f'Relationship {sensor} with PT08.S1(CO)')
        plt.xlabel(sensor)
        plt.ylabel('PT08.S1(CO)')
        
    plt.tight_layout()
    plt.show()
    
sensors = ['PT08.S2(NMHC)', 'NOx(GT)', 'PT08.S3(NOx)', 'PT08.S4(NO2)', 'PT08.S5(O3)','PT08.S1(CO)']
plot_scatter_separately(df=df, target_variable='CO(GT)', sensors=sensors)
````

### VIF (عامل تورم واریانس) به زبان ساده:
**VIF** یه عدده که بهت میگه چقدر **ویژگیهای مستقل** مدلت (مثلاً Xها) به هم شبیهن!  

- **مثال:** فرض کن دوستات تو امتحان همدیگه رو کپی میکنن. اگه ازشون بپرسی "چرا این جواب رو دادی؟" هر دو یه حرف میزنن. اینجا اطلاعات تکراری میشه!  

---

### تفسیر VIF:
- **۱ = عالی** (هیچ شباهتی بین ویژگیها نیست)  
- **۱ تا ۵ = قابل قبول**  
- **۵ تا ۱۰ = مشکوک** (یه کم شبیه هم شدن)  
- **بالای ۱۰ = خطرناک** (ویژگیها خیلی شبیه هم هستند، مثل دوقلوها!)  

---

### چرا مهمه؟  
وقتی ویژگیها **خیلی شبیه هم** باشن:  
- مدل گیج میشه کدوم واقعاً اثرگذاره!  
- نتایج مدل غیرقابل اعتماد میشه!  

--- 

### راهحل:  
اگه VIF بالای ۱۰ بود، یکی از ویژگیهای مشابه رو حذف کن!

````python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculate_vif(X):
    vif = pd.DataFrame()
    vif["Feature"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif

# محاسبه VIF برای گزینه ۲
features_option2 =  [
    'C6H6(GT)',     
    'PT08.S1(CO)',  
    'PT08.S5(O3)',  
    'NOx(GT)',     
    'PT08.S3(NOx)'  
]
vif_result = calculate_vif(df[features_option2])
print(vif_result)
````

````python
# TODO : کتابخانه های مورد نیاز را ایمپورت کنید

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor 
from sklearn.svm import SVR
````

````python
# TODO: چهار مدل اولیه را آموزش دهید و آنها را ارزیابی کنید
# لیست مدلها
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(random_state=101),
    "Random Forest": RandomForestRegressor(random_state=101),
    "Gradient Boosting": GradientBoostingRegressor(random_state=101),
    "SVR": SVR(),
    "XGBoost": XGBRegressor(random_state=101, verbosity=0)
}

# دیکشنری برای ذخیره نتایج
results = {}

# آموزش و ارزیابی هر مدل
for name, model in models.items():
    # آموزش مدل
    model.fit(X_train, y_train)
    
    # پیشبینی روی دادههای آموزش و آزمون
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    # محاسبه معیارها
    train_mae = mean_absolute_error(y_train, y_train_pred)
    train_mse = mean_squared_error(y_train, y_train_pred)
    train_r2 = r2_score(y_train, y_train_pred)
    
    test_mae = mean_absolute_error(y_test, y_test_pred)
    test_mse = mean_squared_error(y_test, y_test_pred)
    test_r2 = r2_score(y_test, y_test_pred)
    
    # ذخیره نتایج
    results[name] = {
        "Train MAE": train_mae,
        "Train MSE": train_mse,
        "Train R²": train_r2,
        "Test MAE": test_mae,
        "Test MSE": test_mse,
        "Test R²": test_r2
    }

# نمایش نتایج
for model, metrics in results.items():
    print(f"\n{model}:")
    print(f"Train R²: {metrics['Train R²']:.3f} | Test R²: {metrics['Test R²']:.3f}")
    print(f"Train MAE: {metrics['Train MAE']:.3f} | Test MAE: {metrics['Test MAE']:.3f}")
````

````python
# TODO : نمودار Learning Curve هر مدل را رسم کنید
from sklearn.model_selection import learning_curve
import numpy as np

def plot_learning_curve(model, X, y, cv=5):
    train_sizes, train_scores, test_scores = learning_curve(
        model, X, y, cv=cv, scoring='r2', 
        train_sizes=np.linspace(0.1, 1.0, 5),
        n_jobs=-1
    )
    
    plt.figure(figsize=(10, 6))
    plt.plot(train_sizes, np.mean(train_scores, axis=1), 'o-', label="Train Score")
    plt.plot(train_sizes, np.mean(test_scores, axis=1), 'o-', label="Test Score")
    plt.title(f"Learning Curve ({model.__class__.__name__})")
    plt.xlabel("Training Examples")
    plt.ylabel("R² Score")
    plt.legend()
    plt.grid(True)
    plt.show()

# رسم نمودار برای مدلهای  (مثلاً Gradient Boosting و Random Forest)
plot_learning_curve(RandomForestRegressor(), X_train, y_train)
plot_learning_curve(GradientBoostingRegressor(), X_train, y_train)
plot_learning_curve(LinearRegression(),X_train,y_train)
plot_learning_curve(XGBRegressor(),X_train,y_train)
````

````python
# TODO: فراخانی کتابخانه های مورد نظر
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import RandomizedSearchCV
````

````python

# دیکشنری برای مشخص کردن فضای جستجوی پارامتر
param_dist_gbr = {
    'n_estimators': [20,50,100],
    'learning_rate': [0.01, 0.05, 0.07],
    'max_depth': [2, 4],
    'min_samples_split': [3,5, 7],
    'min_samples_leaf': [2,3,6,None],
    'subsample': [0.5, 0.7, 0.9],
    'max_features': ['sqrt', 'log2', None],
    'loss': ['squared_error', 'huber', 'absolute_error']
}

# برای GradientBoostingRegressor
gbr_model = GradientBoostingRegressor()

# ایجاد شی از کلاس RandomizedSearchCV
random_search_gbr = RandomizedSearchCV(
    estimator=gbr_model,
    param_distributions=param_dist_gbr,
    n_iter=50,
    scoring="neg_mean_squared_error",
    cv=5,
    verbose=3,
    n_jobs=-1,
    random_state=42
)

# اجرای جستجو (فقط روی Train)
random_search_gbr.fit(X_train, y_train)

# ارزیابی دستی بهترین مدل روی Test
best_model = random_search_gbr.best_estimator_
best_model.fit(X_train, y_train)  # بازآموزی روی کل Train
test_score = best_model.score(X_test, y_test)

print(f'''
{'='*60}
بهترین پارامترها: {random_search_gbr.best_params_}
Test R²: {test_score:.4f}
train r2: {best_model.score(X_train, y_train):.4f}
{'='*60}
''')
````
